```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(color_codes = True)
import matplotlib.pyplot as plt
```

```python
Infections= pd.read_csv('COVID_19_INFECTIONS_AGE_GROUP.csv')
```

```python
Infections.head()
```

```python
Infections.info()
```

```python
sns.set(rc={'figure.figsize':(12.8,9.28)})
sns.stripplot(Infections['Age'], Infections['Infections_%']).set_title("Most affected age groups for July 2021 in %")
```

```python
sns.set(rc={'figure.figsize':(12.8,9.28)})
sns.stripplot(Infections['Age'], Infections['Total']).set_title("Most affected age groups for July 2021 in thousands")
```

```python
sns.set(rc={'figure.figsize':(12.8,9.28)})
sns.barplot(Infections['Age'],Infections['Total']).set_title("Most affected age groups for July 2021 in thousands")
```
