```python
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import plotly.express as px
from jupyter_dash import JupyterDash
from dash.dependencies import Input, Output
```

```python
df = pd.read_csv('COVID_19_CASES_New.csv', sep=',')
df.tail()
```

```python
df.shape
```

```python
# Build App
app = JupyterDash(__name__)
app.layout = html.Div([
    html.H1("Namibia Covid-19 Web Application"),
    dcc.Graph(id='graph'),
    html.Label([
        "Select Region",
        dcc.Dropdown(
            id='Region-dropdown', clearable=False,
            value = "Khomas",options=[
            {'label': 'Khomas', 'value': 'Khomas'},
            {'label': 'Erongo', 'value': 'Erongo'},
            {'label': 'Kharas', 'value': 'Kharas'},
            {'label': 'Hardap', 'value': 'Hardap'},
            {'label': 'Kavango East', 'value': 'Kavango East'},
            {'label': 'Kavango West', 'value': 'Kavango West'},
            {'label': 'Oshana', 'value': 'Oshana'},   
            {'label': 'Omusati', 'value': 'Omusati'},
            {'label': 'Otjozondupa', 'value': 'Otjozondupa'},   
            {'label': 'Oshikoto', 'value': 'Oshikoto'},
            {'label': 'Zambezi', 'value': 'Zambezi'}    
            ])
    ]),
    html.Label([
        "Select Field",
        dcc.Dropdown(
            id='column-dropdown', clearable=False,
            value = "Cases",options=[
            {'label': 'Cases', 'value': 'Cases'},
            {'label': 'Recoveries', 'value': 'Recoveries'},
            {'label': 'Deaths', 'value': 'Deaths'},
            {'label': 'Deaths per day', 'value': 'Deaths per day'},
            {'label': 'Cases per day', 'value': 'Cases per day'},
            {'label': 'Recoveries per day', 'value': 'Recoveries per day'}    
            ])
    ]),
])
# Define callback to update graph
@app.callback(
    Output('graph', 'figure'),
    [Input("Region-dropdown", "value")],
    [Input("column-dropdown", "value")]
)   
def update_figure(region,column):
    return px.scatter(
        df.loc[df['Region'] == region] ,x="Date", y=column,
        render_mode="webgl", title="Scatter Plot"
    )
# Run app and display result inline in the notebook
app.run_server(mode='Inline')
```

```python
# Build App
app = JupyterDash(__name__)
app.layout = html.Div([
    html.H1("Namibia Covid-19 Web Application"),
    dcc.Graph(id='graph'),
    html.Label([
        "Select Field",
        dcc.Dropdown(
            id='Region-dropdown', clearable=False,
            value = "Khomas",options=[
            {'label': 'Khomas', 'value': 'Khomas'},
            {'label': 'Erongo', 'value': 'Erongo'},
            {'label': 'Kharas', 'value': 'Kharas'},
            {'label': 'Hardap', 'value': 'Hardap'},
            {'label': 'Kavango East', 'value': 'Kavango East'},
            {'label': 'Kavango West', 'value': 'Kavango West'},
            {'label': 'Oshana', 'value': 'Oshana'},   
            {'label': 'Omusati', 'value': 'Omusati'},
            {'label': 'Otjozondupa', 'value': 'Otjozondupa'},   
            {'label': 'Oshikoto', 'value': 'Oshikoto'},
            {'label': 'Zambezi', 'value': 'Zambezi'}    
            ])
    ]),
])
# Define callback to update graph
@app.callback(
    Output('graph', 'figure'),
    [Input("Region-dropdown", "value")]
)   
def update_figure(Column):
    return px.line(
        df.loc[df['Region'] == Column] , x="Date",y=["Cases per day","Cases","Recoveries","Recoveries per day","Deaths","Deaths per day"],title="Line Graph,",markers=True
    )
# Run app and display result inline in the notebook
app.run_server(mode='Inline')
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Bar(
    x=df["Region"],
    y=df["Cases"],
    name='Cases',
    marker_color='blue'
))
fig.add_trace(go.Bar(
    x=df["Region"],
    y=df["Cases per day"],
    name='Cases per day',
    marker_color='lightblue'
))
fig.add_trace(go.Bar(
    x=df["Region"],
    y=df["Deaths"],
    name='Deaths',
    marker_color='yellow'
))
fig.add_trace(go.Bar(
    x=df["Region"],
    y=df["Deaths per day"],
    name='Deaths per day',
    marker_color='lightyellow'
))
fig.add_trace(go.Bar(
    x=df["Region"],
    y=df["Recoveries"],
    name='Recoveries',
    marker_color='seagreen'
))
fig.add_trace(go.Bar(
    x=df["Region"],
    y=df["Recoveries per day"],
    name='Recoveries per day',
    marker_color='lightseagreen'
))
# Here we modify the tickangle of the xaxis, resulting in rotated labels.
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show(mode='External')
```
